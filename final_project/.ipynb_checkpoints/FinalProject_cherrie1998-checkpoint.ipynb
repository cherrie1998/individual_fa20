```python
parks_loc = parks_loc[['alias', 'gis_acres']]
parks_loc.head()
```

```python
#From here I know that there's no missing values
rate_null = parks_yelp['rating'].isnull().any()
name_null = parks_yelp['name'].isnull().any()
review_null = parks_yelp['review_count'].isnull().any()
alia_null = parks_loc['alias'].isnull().any()
size_null = parks_loc['gis_acres'].isnull().any()
rate_null or name_null or review_null or alia_null or size_null
```

```python
# As it can be see through the histogram of 0 - 95 percentile of the size is very right skewed, 
# and thus I decide a log transformation for it 
g1 = parks_loc[parks_loc['gis_acres'] <parks_loc['gis_acres'].quantile(0.95)]['gis_acres'].hist()
plt.title('histogram for park size of parks around San Diego')
plt.xlabel('park size')
plt.ylabel('frequency')
```

```python
# As it can be see through the histogram of 0 - 95 percentile of the review counts is very right skewed, 
# and thus I decide a log transformation for it 
g2 = parks_yelp[parks_yelp['review_count'] <parks_yelp['review_count'].quantile(0.95)]['review_count'].hist()
plt.title('histogram for popularities of parks around San Diego')
plt.xlabel('review counts')
plt.ylabel('frequency')
```

```python
parks_loc['gis_acres_log'] = np.log(parks_loc['gis_acres'])
```

```python
parks_yelp['review_count_log'] = np.log(parks_yelp['review_count'])
```

As I want to find a relationship between size and rating, and popularities and rating, I used a linear model to predict the relationship between the variables.

#### This is the model analyzing predicting rating using the size of the park

```python
outcome_1, predictors_1 = patsy.dmatrices('rating ~ gis_acres_log',data = parks_size)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
```

```python
print(res_1.summary())
```

```python
sns.scatterplot(x='gis_acres_log', y='rating', data=parks_size)

# Generate and plot the model fit line
xs = np.arange(parks_size['gis_acres_log'].min(), parks_size['gis_acres_log'].max())
ys = 0.0466 * xs + 3.9116
plt.plot(xs, ys, '--k', linewidth=4, label='Model')

plt.xlabel('gis_acres_log')
plt.ylabel('rating')
plt.legend();
```

##### From the above summary, we can see that the p value of significance is 0.051, so we fail to declare we can predict the rating by the size of the park at significance level 0.01. Although it cannot prove that there's no relationship between size of park and its rating. The above graph is the linear prediction line I got through the analysis. 

#### This is the model analyzing predicting rating using the popularity (number of reviews) of the park

```python
outcome_1, predictors_1 = patsy.dmatrices('rating ~ review_count_log',data = parks)
mod_1 = sm.OLS(outcome_1, predictors_1)
res_1 = mod_1.fit()
```

```python
print(res_1.summary())
```

```python
sns.scatterplot(x='review_count_log', y='rating', data=parks)

# Generate and plot the model fit line
xs = np.arange(parks['review_count_log'].min(), parks['review_count_log'].max())
ys = -0.0109 * xs + 4.0710
plt.plot(xs, ys, '--k', linewidth=4, label='Model')

plt.xlabel('review count')
plt.ylabel('rating')
plt.legend();
```

##### From the above summary, we can see that the p value of significance is 0.41, so we fail to declare we can predict the rating by the number of reviews of park at significance level 0.01. Although it cannot prove that there's no relationship between size of park and its rating. The above graph is the linear prediction line I got through the analysis. 

From the exploratory data analysis, we can see that through the original distribution, when popularity and size get higher, the rate tend to go higher. However, through the linear prediction, we fail to predict the ratings using either the number of reviews of the park or the size of the park.
